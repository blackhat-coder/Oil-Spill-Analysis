```python
def parse_lga(x):
    
    x = x.lower()
    
    for i in x:
        if i in [chr(j) for j in range(33, 48)]:
            x = x.replace(i, ' ')
    
    for i in range(0 , len(x)-2):
        j = i + 1
        if (x[i] == ' ') and (x[j] == ' '):
            x = x.replace(x[j] , '', 1)

    return x


data['LGA'] = data['LGA'].apply(parse_lga)

```

```python
def prep_contaminants(x):
    x = x.lower()
    for i in x:
        if i in [chr(j) for j in range(33, 48)]:
            x = x.replace(i, ' ')
        if i == ':':
            x = x.replace(i, ' ')
    
    for i in range(0 , len(x)-2):
        j = i + 1
        if (x[i] == ' ') and (x[j] == ' '):
            x = x.replace(x[j] , '', 1)

    l = x.split(' ')
    if (len(l) > 1) & (l[0] == 'other'):
        l.pop(0)
        x = ' '.join(l)
    
    return x

data['Contaminant'] = data['Contaminant'].apply(prep_contaminants)

data['Contaminant'].value_counts().keys()
```

```python
# Understand and preprocess contaminants group

# gas is not gs

contaminant_grouping = {
    'oily wt': 'water', 'oily water':'water', 'water from oil pipeline':'water', 'produce water':'water', 'water leak':'water',
    'burnt gas':'gas', 'gaseous emision':'gas',
    'boom': 'damage', 'locally constructed boom':'damage', 'asset damage':'damage',
    'na':'none', 'nil':'none',
    'crude_oil':'crude',
    'none leak':'non leaking',
    ' ':'empty',
    '':'empty'
}

data['New_contaminants'] = data['Contaminant'].map(contaminant_grouping).fillna(data['Contaminant'])

data.New_contaminants.value_counts().index
```
