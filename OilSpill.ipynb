```python
data['Incident_date'] = pd.to_datetime(data['Incident_date'], errors='coerce')
```

```python
data['year'] = data['Incident_date'].dt.year
data['month'] = data['Incident_date'].dt.month
data['dayofweek'] = data['Incident_date'].dt.dayofweek
data['weekofyear'] = data['Incident_date'].dt.week
data['year_quater'] = data['Incident_date'].dt.quarter
data['day'] = data['Incident_date'].dt.day
data['hour'] = data['Incident_date'].dt.hour

```

```python
plt.figure(figsize=(20, 9))
sns.set_style('whitegrid')
sns.set_color_codes(palette="deep")
plt.xticks(rotation=90)

sns.countplot(data=data, x='Spill_area')
```

```python
temp = pd.DataFrame(data.groupby('Spill_area')['Longitude', 'Latitude'].mean())
```

```python
spill_area_map = {}

print(spill_area_map)
print('')


for i, j in zip(data['Spill_area'].value_counts().keys() , data['Spill_area'].value_counts().values):
    spill_area_map[i] = j

print(spill_area_map)
```

```python
temp['density'] = temp.index.map(spill_area_map)
temp
```

```python
sam = folium.Map(location=[6.853860, 4.707157])

for i in range(temp.shape[0]):
    # print(f"{i}    :   {temp.iloc[i]['Longitude']} ")
    long = temp.iloc[i]['Longitude']

    lat = temp.iloc[i]['Latitude']

    r = temp.iloc[i]['density']

    folium.Circle((lat, long),  radius=r, fill=True, color='red', ffill_color='red').add_to(sam)


n = []

for i in range(data.shape[0]):
    if data.loc[i, 'LGA'] not in n:
        lon = data.loc[i,'Longitude']
        lat = data.loc[i,'Latitude']
        # folium.Marker(location=(lat, lon)).add_to(m)
        folium.Circle(location=(lat, lon), radius=100, color="yellow", fill=True, ffil_color="red").add_to(sam)

        n.append(data.loc[i, 'LGA'])

sam
```
